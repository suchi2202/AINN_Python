**Problem statement 2:** AND Gate Scenario for Switching an AC On or Off
John switches on the AC in his room only if both conditions are true: it is hot and the room is occupied. There are four scenarios:



Scenario 1: It is not hot, and the room is not occupied.


Scenario 2: It is not hot, but the room is occupied.


Scenario 3: It is hot, but the room is not occupied.


Scenario 4: It is hot, and the room is occupied.



**SOFTWARE REQUIRED:** Google Colaboratory


**THEORY:**
The McCulloch-Pitts neuron was the earliest neural network discovered in 1943. It is usually called as
M-P neuron. The M-P neurons are connected by directed weighted paths. It should be noted that the activation of a M-P neuron is binary, that is, at any time step the neuron may fire or may not fire.The weights associated with the communication links may be excitatory (weight is positive) or inhibitory
(weight is negative). All the excitatory connected weights entering into a particular neuron will have same weights. The threshold plays a major role in M-P neuron: There is a fixed threshold for each neuron, and if the net input to the neuron is greater than the threshold then the neuron fires. Also, it should be noted that any nonzero inhibitory input would prevent the neuron from firing. The M-P neurons are most widely used in the case of logic functions.
Since the firing of the output neuron is based
upon the threshold, the activation function
here is defined as


f(y) = 1 if y&gt;=0,


f(y) = 0 if y&lt;0..

```python
#AND Gate
import matplotlib.pyplot as plt

class MPNeuron:
    def __init__(self, weights, threshold):
        self.weights = weights
        self.threshold = threshold

    def activation(self, net_input):
        return 1 if net_input >= self.threshold else 0

    def compute_output(self, inputs):
        net_input = sum(w * i for w, i in zip(self.weights, inputs))
        return self.activation(net_input)

# Define the weights and threshold for the AND gate scenario
weights = [1, 1]  # Both conditions (Hot and Occupied) have a weight of 1
threshold = 2     # Threshold is 2, so the neuron fires only if both conditions are true

# Create a neuron for John's AC decision
ac_neuron = MPNeuron(weights, threshold)

# Test scenarios: [Hot, Occupied]
test_scenarios = [
    [0, 0],  # Scenario 1: It is not hot, and the room is not occupied. (Expect 0)
    [0, 1],  # Scenario 2: It is not hot, but the room is occupied. (Expect 0)
    [1, 0],  # Scenario 3: It is hot, but the room is not occupied. (Expect 0)
    [1, 1],  # Scenario 4: It is hot, and the room is occupied. (Expect 1)
]

# Compute the outputs for each scenario
outputs = [ac_neuron.compute_output(scenario) for scenario in test_scenarios]

# Plotting the results
x_values = [1, 2, 3, 4]
y_values = outputs

plt.plot(x_values, y_values, linestyle='-')
plt.xticks(x_values, ["Scenario 1", "Scenario 2", "Scenario 3", "Scenario 4"])
plt.xlabel("Scenarios")
plt.ylabel("Output (0 or 1)")
plt.axhline(y=0, color='gray', linestyle='--', linewidth=0.5)
plt.axhline(y=1, color='gray', linestyle='--', linewidth=0.5)
plt.legend(["Output"])
plt.title("AC Decision Based on Conditions (AND Gate)")
plt.show()
```
