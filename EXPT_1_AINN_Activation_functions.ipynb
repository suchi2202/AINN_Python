```python
import numpy as np
import matplotlib.pyplot as plt

# Sigmoid Activation Function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Tanh Activation Function
def tanh(x):
    return (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x))


# ReLU Activation Function
def relu(x):
    return np.maximum(0, x)

# Leaky ReLU Activation Function
def leaky_relu(x, alpha=0.1):
    return np.where(x > 0, x, alpha * x)

# Softmax Activation Function
'''def softmax(x):
    exp_x = np.exp(x - np.max(x))
    return exp_x / exp_x.sum(axis=0)'''

def softmax(z):
  '''Return the softmax output of a vector.'''
  exp_z = np.exp(z)
  sum = np.sum(np.exp(z))
  softmax_z = exp_z/sum
  return softmax_z

# Swish Activation Function
def swish(x):
    return x * sigmoid(x)

# ELU Activation Function
def elu(x, alpha=1.0):
    return np.where(x > 0, x, alpha * (np.exp(x) - 1))

# Generate values from -10 to 10
x = np.linspace(-10, 20, 400)

# Calculate the activation values
sigmoid_values = sigmoid(x)
tanh_values = tanh(x)
relu_values = relu(x)
leaky_relu_values = leaky_relu(x)
softmax_values = softmax(x)
swish_values = swish(x)
elu_values = elu(x)

# Plot the activation functions
plt.figure(figsize=(8, 12))

plt.subplot(4, 2, 1)
plt.plot(x, sigmoid_values, label='Sigmoid')
plt.title('Sigmoid Activation Function')
plt.grid(True)

plt.subplot(4, 2, 2)
plt.plot(x, tanh_values, label='Tanh', color='r')
plt.title('Tanh Activation Function')
plt.grid(True)

plt.subplot(4, 2, 3)
plt.plot(x, relu_values, label='ReLU', color='g')
plt.title('ReLU Activation Function')
plt.grid(True)

plt.subplot(4, 2, 4)
plt.plot(x, leaky_relu_values, label='Leaky ReLU', color='m')
plt.title('Leaky ReLU Activation Function')
plt.grid(True)

plt.subplot(4, 2, 5)
plt.plot(x, softmax_values, label='Softmax', color='c')
plt.title('Softmax Activation Function')
plt.grid(True)

plt.subplot(4, 2, 6)
plt.plot(x, swish_values, label='Swish', color='y')
plt.title('Swish Activation Function')
plt.grid(True)

plt.subplot(4, 2, 7)
plt.plot(x, elu_values, label='ELU', color='b')
plt.title('ELU Activation Function')
plt.grid(True)

plt.tight_layout()
plt.show()
```

Binary Sigmoid Activation Function: Binary Sigmoid is essential for tasks requiring clear binary decisions, enabling neural networks to excel in binary classification problems. Characterized by its binary output behaviour, it can be prominently used in scenarios requiring binary classification, where it efficiently models decision boundaries and has proven effective in various pattern recognition tasks.

Bipolar Sigmoid Activation Function: Bipolar Sigmoid functions ability to capture both positive and negative patterns make it particularly valuable in tasks where input features exhibit varying polarities. This activation function, distinguished by its bipolar output, offers a versatile tool for modelling complex relationships within data.

ReLU (Rectified Linear Unit) Activation Function: The Rectified Linear Unit (ReLU) activation function is crucial in deep learning for its simplicity, computational efficiency, and ability to address the vanishing gradient problem. ReLU adds non-linearity to neural networks, making them capable of learning complex patterns. Its efficiency speeds up training, allowing for deeper networks. The sparse activation it produces can aid in preventing overfitting. ReLU has a strong track record in various applications, including image recognition and natural language processing, making it a foundational component in modern neural network architectures.

Leaky ReLU Activation Function: Leaky ReLU enhances training stability by addressing the vanishing gradient problem, allowing for smoother optimization and better network performance in deeper architectures. The Leaky ReLU activation function highlights its role in addressing the limitations of the standard ReLU.

```python
class MPNeuron:
    def __init__(self, weights, threshold):
        self.weights = weights
        self.threshold = threshold

    def activation(self, net_input):
        return 1 if net_input >= self.threshold else 0

    def compute_output(self, inputs):
        net_input = sum(w * i for w, i in zip(self.weights, inputs))
        return self.activation(net_input)

# Define the weights and threshold for the AND gate scenario
weights = [1, 1]  # Both conditions (Hot and Occupied) have a weight of 1
threshold = 2     # Threshold is 2, so the neuron fires only if both conditions are true

# Create a neuron for John's AC decision
ac_neuron = MPNeuron(weights, threshold)

# Get input from the user for a single scenario
print("Enter the scenario for the AND gate (1 for true, 0 for false):")
hot = int(input("Is it hot? (1 for Yes, 0 for No): "))
occupied = int(input("Is the room occupied? (1 for Yes, 0 for No): "))

# Create the test scenario and description
test_scenario = [hot, occupied]
description = f"Scenario: Hot = {hot}, Occupied = {occupied}"

# Compute the output for the given scenario
output = ac_neuron.compute_output(test_scenario)
print(f"{description} -> Output: {output}")

# Plotting the result with the scenario description
plt.bar([1], [output], width=0.4)
plt.xticks([1], [description])
plt.ylabel("Output (0 or 1)")
plt.ylim(-0.5, 1.5)
plt.axhline(y=0, color='gray', linestyle='--', linewidth=0.5)
plt.axhline(y=1, color='gray', linestyle='--', linewidth=0.5)
plt.title(f"AC Decision Based on Conditions\n{description}")
plt.show()
```
